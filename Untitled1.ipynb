```python
## Removing Unixstring attached to timestramp 
jira[['timestramp','unixstring']] = jira['time_string'].str.split(',',expand=True)
jira = jira.drop(columns=['time_string', 'unixstring'])

## Casting datetime format 
jira['timestramp']= pd.to_datetime(jira['timestramp']) 

## Creating date & time 
jira['date'] = jira['timestramp'].dt.date
jira['time'] = jira['timestramp'].dt.time
```

## Q2 Daily Active Users Trend

#### From the daily active users trend below two specific things stand out:
#### 1) Users are active 4 days of a week and on one day of each week(probably Sunday ?) very few users are active. 
#### 2) In Last week of December 2014 we see comparitively lower number of users active. This might be due to holiday shutdown being observed by some of our clients.  

```python
sns.set_style("darkgrid")

plt.figure(figsize=(20,10)) 

plt.plot(jira.groupby('date')['username'].nunique(),color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12)

plt.xlabel('Date',fontsize=18) 
plt.ylabel('Active Users',fontsize=18)
plt.title('Daily active Jira Users',fontsize=18)
plt.show() 
```

## Average Active Users On Each Day Of A Week

#### From the plot below we can further confirm that Sunday is the least active day.

```python
new_frame = jira.groupby('date')['username'].nunique().to_frame().reset_index()
new_frame.date = pd.to_datetime(new_frame['date'])

new_frame['day'] = new_frame['date'].dt.day_name().astype(str).str[0:3]

days = new_frame.groupby('day')['username'].mean().to_frame().reset_index().sort_values('username')
plt.bar(days.day, days.username, align='center', alpha=0.5,color=(0.3, 0.3, 0.3, 0.3))
plt.xlabel('Day of week',fontsize=18) 
plt.ylabel('Average Active Users',fontsize=18)
plt.title('Average active Users on each day of week',fontsize=18)
plt.show() 
```

## Q3 Comparing two users

#### By comparing two given users we can see that user "778251ffec35de6bbbb089198b9ac516"  has on average ~123 more events per day than user "7e7a085a9f6f55253b2aa15dbcb702a8".

```python
user_one = jira[(jira['username']== '778251ffec35de6bbbb089198b9ac516') & (jira['instance']=='c93da7e17688ef90fc43963b5a6c6259')].groupby(['date']).size().to_frame().reset_index()
user_one.columns = ['date','516 User']
user_two = jira[(jira['username']== '7e7a085a9f6f55253b2aa15dbcb702a8') & (jira['instance']=='c93da7e17688ef90fc43963b5a6c6259')].groupby(['date']).size().to_frame().reset_index()
user_two.columns = ['date','2a8 User']
users = pd.merge(user_one,user_two, on ='date')

user1 = users['516 User'].mean()
user2 = users['2a8 User'].mean()

print("The average number of events per day for 778251ffec35de6bbbb089198b9ac516 is ", user1 ,"and the average number of events per day for 7e7a085a9f6f55253b2aa15dbcb702a8 is ", user2, ".")
```

#### When we look at daily usage trend for these two users we can further confirm that their usage pattern isn't exactly the same. Both users have peaks and lows at differnt intervals. 

```python
plt.figure(figsize=(20,10)) 
plt.plot(users['date'], users['516 User'], color='g',  label='516 User')
plt.plot(users['date'], users['2a8 User'], color='orange', label = '2a8 User')
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Events per day', fontsize = 18)
plt.legend(loc='upper left')

plt.title('Count of Events for two users', fontsize = 18)

plt.show()
```
